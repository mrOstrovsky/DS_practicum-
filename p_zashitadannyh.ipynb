Обозначения:

- $X$ — матрица признаков (нулевой столбец состоит из единиц)

- $y$ — вектор целевого признака

- $P$ — матрица, на которую умножаются признаки

- $w$ — вектор весов линейной регрессии (нулевой элемент равен сдвигу)

Предсказания:

$$
a = Xw
$$

Задача обучения:

$$
w = \arg\min_w MSE(Xw, y)
$$

Формула обучения:

$$
w = (X^T X)^{-1} X^T y
$$

**Ответ:** Не изменится.

**Обоснование:** 

**1. У нас имеется обратимая(квадратная) матрица**

$$
P
$$

**2. Зашифруем Х**

$$
X = C

**3. Проведем регерессию для новой матрицы признаков (по стандартной формуле)**
$$
w = (X^T X)^{-1} X^T y
$$

Введем новое значение для весов для **С**

$$
u
$$

$$
Cu,y => u = (C^T C)^{-1} C^Ty = ((XP)^{T}XP)^{-1}(XP)^{T}y
$$

Мы значем принципы вычисления:

$$ 
(AB)^{T} = B^{T}A^{T}
$$

$$ 
(AB)^{-1} = B^{-1} A^{-1}
$$


**4. Преобразуем матрицу:**


$$
((XP)^{T}XP)^{-1}(XP)^{T}y = (P^T X^T XP)^{-1} P^T X^T y 
$$

Мы знаем что в формуле $$w = (X^T X)^{-1} X^Ty$$ есть пара произведений $$(X^T X)$$ 

Для удобства в дальнейшем преобразовании добавим в скобки $$X^T X => (P(X^T X)P)^{-1}P^T X^T y$$ 

Продолжаемм преобразование:

$$(P(X^T X)P)^{-1}P^T X^T y = P^{-1}(X^T X)^{-1} (P^T)^{-1}P^T X^T y = P^{-1}(X^T X)^{-1}X^Ty =>  P^{-1}w$$

Получаем: $$u = P^{-1}w$$

**5. Узнаем чему равна *w* для С**

Мы знаем что **u** это вес для **С**, а **w** вес для **Х**, чтобы не считать регрессию заново, мы можем восстановить(расшифровать) чему будет равна **w**, используем след. выражение:



$$u = P^{-1}w$$

w = Pu

## Проверка алгоритма

**1. Создадим переменные с признаками и целевым признаком**

```python
features = df.drop(columns='Страховые выплаты', axis=1)
target = df['Страховые выплаты']
```

**2. Создадим тренировочную и тестовую выборку**

```python
features_train, features_test, target_train, target_test = train_test_split(features, 
                                                                            target, 
                                                                            random_state=12345,
                                                                            shuffle=True,
                                                                            test_size=0.25)
```

**3. Создадим свою модель линейной регрессии без шифрования и посмотрим результат R2**

```python
class LinearRegression:
    def fit(self, features_train, target_train):
        X = np.concatenate((np.ones((features_train.shape[0], 1)), features_train), axis=1)
        y = target_train
        w = np.linalg.inv(X.T.dot(X)).dot(X.T).dot(y)
        self.w = w[1:]
        self.w0 = w[0]

    def predict(self, features_test):
        return features_test.dot(self.w) + self.w0
    
model = LinearRegression()
model.fit(features_train, target_train)
predictions = model.predict(features_test)
r2 = r2_score(target_test, predictions)
print('Значение метрики R2 для линейной регрессии без шифрования',r2_score(target_test, predictions))
```

**4. Cоздадим свою зашифрованную модель линейной регрессии и посмотрим на результат R2**

**Функция, которая сгенерирует обратимую матрицу нужного нам размера.**

```python
 def get_rand_matrix(selection):
        det = 0
        while det == 0:
            matrix = np.random.normal(size=(features.shape[1], features.shape[1]))
            det = np.linalg.det(matrix)
                   
        return matrix
```

```python
class encryptedModel:
    
    def fit(self, features_train, target_train):
        self.R = get_rand_matrix(features_train)
        self.P = features_train.dot(self.R)
        X = np.concatenate((np.ones((self.P.shape[0], 1)), self.P), axis=1)
        y = target_train
        w = np.linalg.inv(X.T.dot(X)).dot(X.T).dot(y)
        self.w = w[1:]
        self.w0 = w[0]
        
    def predict(self, features_test):
        return (features_test @ self.R).dot(self.w) + self.w0
    
model = encryptedModel()
model.fit(features_train, target_train)
predictions = model.predict(features_test)
r2crypt = r2_score(target_test, predictions)
print('Значение метрики R2 для линейной регрессии c шифрованием', r2_score(target_test, predictions)) 
```

**5. Сравним результат R2 двух наших моделей линейной регрессии.**

```python
R2_result = pd.DataFrame(index=['R2'],
                                 columns=['Модель без шифрования','Модель с шифрованием'])
R2_result['Модель без шифрования'] = r2
R2_result['Модель с шифрованием'] = r2crypt

R2_result
    
```
