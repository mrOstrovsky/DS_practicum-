-Значения в **year_of_release** поменяет на **int**, чтобы убрать **".0"** после года.

-**tbd (to be determinated)** в **user_score** заменим на **NaN**, так как рейтинг не определен.


## Подготовка данных

```python
# Приводим значения в столбце с пользовательским рейтингом к численному типу данных
# значение tbd (to be determinated) заменяем на пропуски, так как рейтинг не определен
data_games['user_score'] = pd.to_numeric(data_games['user_score'], errors='coerce')
```

```python
# Меняем тип данных на 'int' в 'year_of_release'
data_games['year_of_release'] = data_games['year_of_release'].fillna(0)
data_games['year_of_release'] = data_games['year_of_release'].astype('int', errors='ignore')
```

```python
#Cчитаем суммарные продажи во всех регионах и записываем их в отдельный вектор.
data_games['sum_sales'] = data_games['na_sales'] + data_games['eu_sales'] + data_games['jp_sales'] + data_games['other_sales']
data_games.head(10)
```

### Вывод




-Значения в **'user_score'** приведены к численному типы данных 


-Значения **tbd** заменены на пропуски


-Изменен тип данных в **'year_of_release'** на **int**


-Создан столбец **'sum_sales'** с суммой продаж по всем регионам

## Исследовательский анализ данных

```python
#Смотрим сколько игр выпускалось в разные годы
data_games.plot(y='year_of_release', title='Количество выпусков игр по годам', range=(1995, 2016),
                grid=True, kind='hist', bins=70);
```

Больше всего игр выпускалось с **2005** по **2011** (особенно в **2008** и **2009**).

```python
#Смотрим суммарные продажи по платформам
data_games.groupby('platform')['sum_sales'].sum().sort_values(ascending=False) 
```

```python
#Формируем список платформ с суммарными продажами > 200 млн. долларов 
data_top = (data_games.groupby('platform')
            .agg({'sum_sales': 'sum'})
            .query('sum_sales > 200')
            .sort_values(by='sum_sales', ascending=False)
           )

data_top
```

```python
#Делаем срез по платформам с суммарными продажами > 200 млн. долларов
data_games.query('platform in @data_top.index').head(20)
```

```python
#Формируем сводную таблицу с прибыльными платформами по годам
#Пропуски заменим на 0 
(data_games
 .query('platform in @data_top.index')
 .pivot_table(index='year_of_release', columns='platform', values='sum_sales', aggfunc='sum')
 .fillna(0)
)
```

```python
#Строим распределение продаж на платформах по годам 

(data_games
 .query('platform  in @data_top.index and year_of_release >= 1995')
 .pivot_table(index='year_of_release', columns='platform', values='sum_sales', aggfunc='sum')
 .fillna(0)
 .plot(grid=True, figsize=(15, 5), title='Распределение продаж на платформах по годам с 1995 по 2016 годы')
)
plt.xlabel('Год выпуска игры')
plt.ylabel('Продажи, млн. долларов');
plt.legend(bbox_to_anchor=(1, 1))
```

### Вывод


 
 
-Жизненный цикл платформ состовляет примерно **10 лет**.


-Чтобы сделать прогноз мы возьем половину жизненного цикла платформы.


-Для того чтобы сделать прогноз на **2017 год, мы возьмем срок за актуальный период > 2011 года.**

```python
#Выделяем платформы за актуальный срок
data_games.query('year_of_release > 2011').groupby('platform')['sum_sales'].sum().sort_values(ascending=False)
```

```python
# Распределение для платформ за актуальный период
data_period_grouped = (
    data_games
    .query('year_of_release > 2011')
    .pivot_table(index='year_of_release', columns='platform', values='sum_sales', aggfunc='sum')
    .fillna(0)
)

data_period_grouped.plot(grid=True, figsize=(15, 5), title='Распределение продаж на платформах c 2011 по 2016 года')
plt.xlabel('Год выпуска игры')
plt.ylabel('Продажи, млн. долларов');
```

-Продажи **PS3, X360, Wii, DS, PC, PSP, PS2, WiiU, PSV, XB** и **DC** падают на **2015 год**


-Продажи **PS4, XOne** и **3DS** растут на **2015 год**
Если учитывать, что они только набирают популярность, данные платформы являются потенциально прибыльными.


С **2012** по **2016** год самыми прибыльными были платформы: **PS3, X360 и PS4**


```python
#Строим график «ящик с усами» по глобальным продажам игр в разбивке по платформам 
plt.figure(figsize=(15, 5))
sns.boxplot(data=data_period_grouped, orient='h')
plt.xlabel('Продажи, млн. долларов')
plt.ylabel('Платформа')
plt.title('Диаграммы размаха по глобальным продажам игр в разбивке по платформам');
```

По диаграмме видно, что на платформе **Wii** встречаются самые дорогие продажи, но медиана смещена влево, значит они встречаются редко. На платформе **X360**, медиана смещена вправо, и она больше среднего, значит на данной платформе много продаж со средней суммой в 100 млн. долларов. **На платформе PS3, которая лидирует по суммарным продажам игр**, медиана смещена еще правее, значит на ней много продаж выше среднего.

```python
#Формируем новую таблицу, в которой не будем учитывать данные за предыдущие годы
data_games_period = data_games.query('year_of_release > 2011')
data_games_period.head(10)
```

**PS3**

```python
#Делаем срез по популярной платформе PS3
data_PS3 = data_games_period.query('platform == "PS3"')
data_PS3.head(10)
```

```python
#Узнаем, как влияют на продажи внутри платформы PS3 отзывы пользователей, находим корреляцию
data_PS3['sum_sales'].corr(data_PS3['user_score'])
```

```python
#Узнаем, как влияют на продажи внутри платформы PS3 отзывы критиков, находим корреляцию
data_PS3['sum_sales'].corr(data_PS3['critic_score'])
```

```python
#Узнаем, как влияют на продажи отзывы пользователей, находим корреляцию
#data_games_period['sum_sales'].corr(data_games_period['user_score'])
```

```python
for platform in ['PS3', 'X360', '3DS', 'PS4']:
    print('\n{0}:'.format(platform))
    for column in ['user_score', 'critic_score']:
        temp = data_games_period.query('platform == @platform')
        print('Коэффициент корреляции {0} и sum_sales: {1}'.\
              format(column, round(temp[column].corr(temp['sum_sales']), 2)))
        temp.plot(x=column, y='sum_sales', kind='scatter', figsize=(15, 5), grid=True) 
        plt.title(platform)
```

**Значения оценок не связаны между собой**

```python
#Строим график зависимости продаж по всем платформам от оценок пользователей
(data_games_period
 .pivot_table(index='user_score', columns='platform', values='sum_sales', aggfunc='sum')
 .plot(figsize=(25, 10), grid=True, title='Распределение продаж по всем платформам в зависимости от оценок пользователей')
)
plt.xlabel('Оценка пользователей')
plt.ylabel('Продажи, млн. долларов');
```

```python
#Строим график зависимости продаж по всем платформам от оценок критиков
(data_games_period
 .pivot_table(index='critic_score', columns='platform', values='sum_sales', aggfunc='sum')
 .plot(figsize=(20, 10), grid=True, title='Распределение продаж по всем платформам в зависимости от оценок критиков')
)
plt.xlabel('Оценка критиков')
plt.ylabel('Продажи, млн. долларов');
```

-Оценки пользователей корреляции с продажами не имеют.


-Оценки критиков демонстрируют прямую, средне-слабую корреляцию с "цифрами" продаж.
**По диаграммам рассеяния и графикам видно, что продажи игр возрастают с ростом оценки.**

```python
#Формируем сводную таблицу игр по жанрам

data_games_period.pivot_table(index='genre', values='sum_sales', aggfunc='median').sort_values(by=['sum_sales'],ascending=False)
```

```python
#Строим распределение продаж по жанрам
(data_games_period
 .pivot_table(index='genre', values='sum_sales', aggfunc='median')
 .sort_values(by='sum_sales')
 .plot(kind='bar', figsize=(15, 5), grid=True, title='распределение продаж по жанрам')
)
plt.xlabel('Жанры')
plt.ylabel('Продажи, млн. долларов');
```

## Проверяем гипотезы

Гипотезы:


1. Средние пользовательские рейтинги платформ **Xbox One и PC** одинаковые;



2. Средние пользовательские рейтинги жанров **Action** (англ. «действие», экшен-игры) и **Sports**(англ. «спортивные соревнования») разные.

```python
#Делаем копию дата фрейма чтобы удалить пропущенные значения для проверки гипотез 
d_g_p = data_games_period.copy()
```

```python
#Cмотрим пропущееные значения
d_g_p.isna().sum()
```

```python
#Удаляем пропущенные значения
d_g_p.dropna(subset=['critic_score'], inplace=True)
d_g_p.dropna(subset=['user_score'], inplace=True)
d_g_p.dropna(subset=['rating'], inplace=True)
```

```python
#убедимся что все удалили
d_g_p.isna().sum()
```

**1.**

```python
#Делаем срез таблицы с играми по платформам PC и XOne
data_games_PC = d_g_p.query('platform == "PC"')
data_games_XOne = d_g_p.query('platform == "XOne"')
```

Так как данные о пользовательских оценках игр представлены только двумя разными выборками, то чтобы проверить гипотезу о равенстве средних пользовательских рейтингов платформ **Xbox One и PC**, сравним их средние значения с учетом дисперсии. Данные выборки независимы между собой, так как в этих оценку давали различные пользователи, распределены нормально, поэтому применим статистический **t-тест**, используя метод **scipy.stats.ttest_ind (array1, array2, equal_var)** для двух выборок.

```python
# Имеем 2 выборки. Проверим гипотезу о равенстве их средних и сформулируем гипотезы так: 
# H0 = Средние пользовательские рейтинги платформ Xbox One и PC одинаковые
# H1 = Средние пользовательские рейтинги платформ Xbox One и PC различаются
# equal_var = False, так как дисперсии выборок не равны
# Критический уровень статистической значимости 0.05
alpha = 0.05


results = st.ttest_ind(data_games_PC['user_score'], data_games_XOne['user_score'], equal_var = False)
print('p-значение: ', results.pvalue)
if (results.pvalue < alpha):
      print("Отвергаем нулевую гипотезу")
else:
      print("Не получилось отвергнуть нулевую гипотезу")
```

**2.**

```python
#Делаем срез таблицы с играми по жанрам Action и Sports
data_games_action = d_g_p.query('genre == "Action"')
data_games_sports = d_g_p.query('genre == "Sports"')
```

Данные о пользовательских оценках игр представлены только двумя разными выборками, то чтобы проверить гипотезу о различии средних пользовательских рейтингов жанров Action и Sports, сравним их средние значения с учетом дисперсии. Данные выборки независимы между собой, так как в этих оценку давали различные пользователи, распределены нормально, поэтому применим статистический тест t-тест, используя метод scipy.stats.ttest_ind (array1, array2, equal_var) для двух выборок. Основную гипотезу возьмем со знаком равенства средних

```python


# H0 = Средние пользовательские рейтинги жанров Action и Sports одинаковые
# H1 = Средние пользовательские рейтинги жанров Action и Sports различаются
# equal_var = False, так как дисперсии выборок не равны
# Критический уровень статистической значимости 0.05
alpha = 0.05


results = st.ttest_ind(data_games_action['user_score'], data_games_sports['user_score'], equal_var = False)
print('p-значение: ', results.pvalue)
if (results.pvalue < alpha):
      print("Отвергаем нулевую гипотезу")
else:
      print("Не получилось отвергнуть нулевую гипотезу")
```
