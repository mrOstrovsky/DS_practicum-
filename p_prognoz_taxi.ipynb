#### **Тест Дики-Фуллера**

Один из способов проверить, является ли временной ряд стационарным, — это выполнить расширенный тест Дики-Фуллера , в котором используются следующие нулевая и альтернативная гипотезы:

**H 0 :** Временной ряд является нестационарным. Другими словами, он имеет некоторую структуру, зависящую от времени, и не имеет постоянной дисперсии во времени.

**H A :** временной ряд является стационарным.

Если **p-значение** из теста меньше некоторого уровня значимости (например, **α = 0,05**), то мы можем отвергнуть нулевую гипотезу и сделать вывод, что временной ряд является стационарным.

4 версии теста:


Без константы и тренда: **'nc'**


С константой, но без тренда: **'c'**


С константой и линейным трендом: **'ct'**


С константой, линейным и квадратичным трендом: **'ctt'**

```python
ts = df['num_orders']
st_test = adfuller(ts, regression='ctt')

if st_test[1] < 0.05:
    print('Ряд стационарный')
else:
    print('Ряд нестационарный')
```

## Обучение

```python
RS=12345
```

### Подготовим признаки

```python
def make_features(data, max_lag, rolling_mean_size):
    data['hour'] = data.index.hour
    data['dayofweek'] = data.index.dayofweek
    
    for lag in range(1, max_lag + 1):
        data['lag_{}'.format(lag)] = data['num_orders'].shift(lag)

    data['rolling_mean'] = data['num_orders'].shift().rolling(rolling_mean_size).mean()
```

**rolling_mean_size** подбирал экспериментально 

```python
make_features(df, 24, 12)
```

### Создадим выборки 

Сделаем тестовую выборку размером 10% от исходных данных

```python
train, test = train_test_split(df, shuffle=False, random_state=RS, test_size=0.1)
train = train.dropna()
 
features_train = train.drop(['num_orders'], axis=1)
target_train = train['num_orders']
features_test = test.drop(['num_orders'], axis=1)
target_test = test['num_orders']
 
```

### Обучение моделей

**RMSE** адекватной модели (методом прогноза предыдущим значением ряда)

```python
train_bl, test_bl = train_test_split(df, shuffle=False, test_size=0.2)

pred_previous = test_bl.shift()
pred_previous.iloc[0] = test_bl.iloc[-1]
print("RMSE:",mean_squared_error(test_bl.num_orders, pred_previous.num_orders)**0.5)
```

#### RandomForestRegressor

```python
%%time

model_rf = RandomForestRegressor()

param_grid_rf = {
    'n_estimators': [10,50],
    'max_depth': [10,50],
    'random_state' : [RS],
}

refit = True
scoring='neg_root_mean_squared_error'
n_jobs = -1
tscv = TimeSeriesSplit(n_splits=5)
gs_rf = GridSearchCV(estimator=model_rf,
                           param_grid=param_grid_rf,
                           refit=refit,
                           scoring=scoring,
                           cv=tscv,
                           n_jobs=n_jobs).fit(features_train, target_train)


display(gs_rf.best_score_ * -1)
display(gs_rf.best_params_)
```

#### DecisionTreeRegressor

```python
%%time

model_dt = DecisionTreeRegressor()

param_grid_dt = {
    'max_depth': [10,50],
    'random_state' : [RS],
}

refit = True
scoring='neg_root_mean_squared_error'
n_jobs = -1
tscv = TimeSeriesSplit(n_splits=5)
gs_dt = GridSearchCV(estimator=model_dt,
                           param_grid=param_grid_dt,
                           refit=refit,
                           scoring=scoring,
                           cv=tscv,
                           n_jobs=n_jobs).fit(features_train, target_train)


display(gs_dt.best_score_ * -1)
display(gs_dt.best_params_)
```

#### LGBMRegressor

```python
%%time

model_lgbm = LGBMRegressor()

param_grid_lgbm = {'boosting_type': ['gbdt'],
                   'n_estimators': [10,50],
                   'max_depth': [10,50],
                   'random_state' : [RS]
                  }

refit = True
scoring='neg_root_mean_squared_error'
n_jobs = -1
tscv = TimeSeriesSplit(n_splits=5)

gr_lgbm = GridSearchCV(estimator = model_lgbm,
                       param_grid = param_grid_lgbm,
                       scoring = scoring,
                       n_jobs = n_jobs,
                       cv=tscv).fit(features_train, target_train)   

display(gr_lgbm.best_params_)
display(gr_lgbm.best_score_*-1)
```

#### **Вывод:**

Привожу таблицу сравнения метрики **RMSE**(оценка проводилась кроссвалидацией) моделей с оптимизированными параметрами.





```python
sravnenie = [[53.42502252385307],
        [25.352687831256336],
        [32.78728364685291],
        [24.945012479916656]]
             
model = ["Адекватная модель","RandomForestRegressor", "DecisionTreeRegressor", "LGBMRegressor"]

pd.DataFrame(data=sravnenie, index=model, columns=["RMSE"])
```

Наилучшей моделью оказалась **LGBMRegressor**

## Тестирование

### Тестирование модели **LGBMRegressor**

```python
model_lgbm = LGBMRegressor(random_state=RS, max_depth=10, n_estimators=50, boosting_type='gbdt')

model_lgbm.fit(features_train, target_train) 
```

```python
model_lgbm_prediction_test = model_lgbm.predict(features_test)

rmse_test = mean_squared_error(target_test, model_lgbm_prediction_test, squared=False)

rmse_test
```
